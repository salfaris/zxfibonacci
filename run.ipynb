```python
from zxfibo import zxfibo, ZXFibo

from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt
```

```python
def fibo(n):
    """Classical Fibonacci algorithm."""
    if n == 0:
        return 1
    if n == 1:
        return 1
    return fibo(n-1) + fibo(n-2)
```

```python
N = 9
print("Quantum:", end=" ")
for i in range(2, N):
    print(zxfibo(i), end=" ")
print()
print("Classical:", end=" ")
for i in range(3, N+1):
    print(fibo(i), end=" ")
```

```python
for i in range(2, N):
    fig, ax = plt.subplots()
    count, _ = ZXFibo(i).eval()
    fig = plot_histogram(count, title=f'Ideal count for ZXFibo n={i}', ax=ax)
    plt.show()
```
